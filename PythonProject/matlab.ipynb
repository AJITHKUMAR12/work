```python
from matplotlib import pylab
```

```python
print(pylab.__version__)
```

```python
import numpy as np
array=np.arange(10)
print(array)
```

```python
x=np.linspace(0,10,10)
print(x)
```

```python
y=x*x+2
print(y)
```

```python
pylab.plot(x,y,'r')
```

```python
pylab.subplot(1,2,1)
pylab.plot(x,y,'r--')
pylab.subplot(1,2,2)
pylab.plot(y,x,'g--')
```

```python
pylab.subplot(1,2,1)
pylab.plot(x,y,'r--')
pylab.subplot(1,2,1)
pylab.plot(y,x,'g--')
```

```python
from matplotlib import pyplot as plt
```

```python
fig = plt.figure()
axis = fig.add_axes([0.5,0.1,0.8,0.8])
axis.plot(x,y,'r')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=1)
for ax in axes:
    ax.plot(x,y,'r')
```

```python
 fig=plt.figure()
 axes2=fig.add_axes([0.1,0.1,0.8,0.8])
 axes3=fig.add_axes([0.2,0.5,0.4,0.3])
 axes2.plot(x,y,'r')
 axes3.plot(x,y,'g')
```

```python
fig=plt.figure(figsize=(8,8))
fig.add_subplot()
plt.plot(y,x,'r')
```
